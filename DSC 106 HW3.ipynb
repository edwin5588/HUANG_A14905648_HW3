```python
import json
import os
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
fp = os.path.join("springfield_converted_json.js")
with open(fp, 'r') as f:
    data = json.load(f)
```

```python
# getting dataframe
df = pd.DataFrame()
for col in range(0, len(data)):
    if data[col]['units'] != "MW":
        continue
    else:
        df[data[col]['id']] = pd.DataFrame(data[col]['history']['data'])[0]
        
        
start = datetime.fromtimestamp(1571579700)
end = datetime.fromtimestamp(1572184500)

times = []
time = start
while time < end:
    times.append(time.strftime("%Y-%m-%d    %X"))
    time += timedelta(minutes = 5)
df["time"] = times
        
df.set_index("time")
```
