```python
import json
import os
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
fp = os.path.join("springfield.json")
with open(fp, 'r') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame()
for col in range(0, len(data)):
    if data[col]['units'] != "MW":
        continue
    else:
        df[data[col]['id']] = pd.DataFrame(data[col]['history']['data'])[0]
df = df.drop("Springfield.fuel_tech.rooftop_solar.power", axis = 1)
```
